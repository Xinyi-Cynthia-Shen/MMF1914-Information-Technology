### Conditionals

As their name indicates, conditionals are a way to execute code depending on whether a condition is True or False. As in other languages, Python supports `if` and `else` but `else if` is contracted into `elif`, as the example below demonstrates.


```

  print("negative")

  print("positive")
```



````python
my_var = -1

if my_var < 0:
  print("less than zero")
elif my_var == 0:
  print("zero")
else:
  print("greater than zero")
````

Here `<` and `>` are the strict `less` and `greater than` operators, while `==` is the equality operator (not to be confused with `=`, the variable assignment operator). The operators `<=` and `>=` can be used for less (resp. greater) than or equal comparisons.

## Discrete Signal, Array and Matrix.
The central data construct in Python is the numeric array, an ordered collection of real or complex numeric data with one or more dimensions. The basic data objects of signal processing (one-dimensional signals or sequences, multichannel signals, and two-dimensional signals) are all naturally suited to array representation.

Python represents ordinary one-dimensional sampled data signals, or sequences, as vectors. Vectors are 1-by-n or n-by-1 arrays, where n is the number of samples in the sequence.

One way to introduce a sequence into python is, to enter it as a list of elements. Create a code box immediately after this text and type the following command.

````python
import pylab as plt

plt.figure(1)
Fs = 100
N = 1000
t1 = np.linspace(0, N, N+1)/Fs # x axis (period, frequency)
x1 = np.sin(2*np.pi*2*t1) # y axis, sin function
plt.plot(t1, x1)

plt.figure(2)
stoptime = 10
t2 = np.linspace(0, stoptime, N+1) # x axis (period, frequency)
x2 = np.sin(2*np.pi*3*t2)  # y axis, sin function
plt.plot(t2, x2)
````

## Task 4:  Creating the Black Schloles Equation



This program is to evaluation the option price using BSM Model, covers both Call and Put Option.

````python
import numpy as np
import scipy.stats as si
import sympy as sy
#import sympy.statistics as systats
````

````python
def euro_vanilla_call(S, K, T, r, sigma,q):
    # assuming q to be 0 and not incorporated in the equation

    #S: spot price
    #K: strike price
    #T: time to maturity
    #r: interest rate
    #sigma: volatility of underlying asset
    #q: dividend yield (usually 0)

    d1 = (np.log(S/K) + (r - q + sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)

    call = si.norm.cdf(d1)*S*np.exp(-q*T) - si.norm.cdf(d2)*K*np.exp(-r*T)

    return call
````

````python
euro_vanilla_call(50, 100, 1, 0.05, 0.25,0)
````

````python
def euro_vanilla_put(S, K, T, r, sigma,q):

    #S: spot price
    #K: strike price
    #T: time to maturity
    #r: interest rate
    #sigma: volatility of underlying asset
    #q: dividend yield (usually 0)

    d1 = (np.log(S/K) + (r - q + sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)

    put = -si.norm.cdf(-d1)*S*np.exp(-q*T) + si.norm.cdf(-d2)*K*np.exp(-r*T)

    return put
````

````python
euro_vanilla_put(50, 100, 1, 0.05, 0.20,0)
````
